```python
def calculator(stage, closing_prices, i, bought,sell):
    print(stage)
    if stage == 0:
        if closing_prices["Close"][i] > closing_prices["EMA13"][i]:
            return 1, bought,sell  # Go to stage 1 and set bought to True
        else:
            return 6,  bought,sell  # Go to stage 6 and set bought to False
    elif stage == 1:
        if closing_prices["High"][i] < closing_prices["Upper_bb"][i]:
            return 0, bought,sell  # Go back to the origin, maintain bought status
        elif closing_prices["Close"][i] <= closing_prices["EMA13"][i]:
            return 2,  bought,sell  # Go to sell opportunity 2, maintain bought status
        else:
            return 1,  bought,sell  # Continue, maintain bought status
    elif stage == 2:
        if closing_prices["High"][i] < closing_prices["Upper_bb"][i]:
            return 0, bought,sell  # Go back to the origin, maintain bought status
        elif closing_prices["Close"][i] <= closing_prices["EMA13"][i]:
            return 2,  bought,sell  # Continue, maintain bought status
        else:
            return 3,  bought,sell  # Go to sell opportunity 3, maintain bought status
    elif stage == 3:
        if closing_prices["High"][i] < closing_prices["Upper_bb"][i]:
            return 0, bought,sell  # Go back to the origin, maintain bought status
        elif closing_prices["Close"][i] <= closing_prices["EMA13"][i]:
            return 4,  bought,sell  # Sell, set bought to False
        else:
            return 3,  bought,sell  # Continue, maintain bought status
    elif stage == 4:
        if (closing_prices["EMA50"][i] - closing_prices["Low"][i]) / closing_prices["Low"][i] < 0.001:
            return 5,  bought,True   # Continue, maintain bought status
        elif closing_prices["EMA50"][i] <= closing_prices["Low"][i]:
            return 5,  bought,True  # Continue, maintain bought status
        else:
            return 6, bought,True   # Go to sell opportunity 3, maintain bought status
    elif stage == 5:
        return 0, bought,sell  # Net zero, maintain bought status
    elif stage == 6:
        if closing_prices["Low"][i] < closing_prices["lower_bb"][i]:
            return 0, bought,sell  # Go back to the origin, maintain bought status
        elif closing_prices["Close"][i] >= closing_prices["EMA13"][i]:
            return 7,  bought,sell  # Go to buy opportunity 2, set bought to True
        else:
            return 6,  bought,sell  # Continue, maintain bought status
    elif stage == 7:
        if closing_prices["Low"][i] < closing_prices["lower_bb"][i]:
            return 0, bought,sell  # Go back to the origin, maintain bought status
        elif closing_prices["Close"][i] <= closing_prices["EMA13"][i]:
            return 8,  bought,sell  # Go to buy opportunity 3, maintain bought status
        else:
            return 7,  bought,sell  # Continue, maintain bought status
    elif stage == 8:
        if closing_prices["Low"][i] < closing_prices["lower_bb"][i]:
            return 0, bought,sell  # Go back to the origin, maintain bought status
        elif closing_prices["Close"][i] >= closing_prices["EMA13"][i]:
            return 9,  bought,sell  # Go to buy opportunity 3, set bought to True
        else:
            return 8,  bought,sell  # Continue, maintain bought status
    elif stage == 9:
        if (closing_prices["High"][i] - closing_prices["EMA50"][i]) / closing_prices["High"][i] < 0.001:
            return 5,  True,sell  # Continue, maintain bought status
        elif closing_prices["EMA50"][i] >= closing_prices["High"][i]:
            return 5,  True,sell  # Continue, maintain bought status
        else:
            return 1,  True,sell  # Go to stage 1, set bought to True
```
